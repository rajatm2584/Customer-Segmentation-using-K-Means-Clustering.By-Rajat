```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Snapshot date: day after last invoice
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# RFM table
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,   #Recency
    'InvoiceNo': 'nunique',                                    # Frequency
    # Monetary: Total spent = Quantity * UnitPrice
    'Quantity': lambda x: (df.loc[x.index,  'Quantity'] * df.loc[x.index, 'UnitPrice']).sum()  #Monetary

})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

### COMPONENT- 4 Run K-means Clustering

## **A. Standardarize the data**

```python
## Standardization

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

##  Standardization is important- Improves model performance and interpretability,and Results become more reliable and comparable.


## **B. Determining optimal no of cluters**

```python
## Determining actual nos of cluster through elbow method- we use WCSS to decide best no of clusters
from sklearn.cluster import KMeans
wcss = []

# Try K from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

## Interpretation: Looking for the "elbow point" where the WCSS drops sharply and then levels off — that's the ideal number of clusters (let's say it's K = 4 for example).

### **C. Run K-means Clustering**

```python
# Apply KMeans with optimal K (say K=4)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(rfm_scaled)

# Assign cluster labels
rfm['Cluster'] = kmeans.labels_
```

# Intertpretation:-
n_clusters=4:
WE're grouping customers into 4 distinct clusters, meaning the optimal number of customer segments has already been determined (using the elbow method, but we could use the silhouette score).

init='k-means++':
A smart way to initialize centroids that helps improve performance and avoid poor clustering due to bad starting points.

random_state=42:
Ensures reproducibility — the clustering results will be the same every time we run this code.
